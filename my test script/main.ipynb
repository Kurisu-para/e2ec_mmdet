```python
import torch, torchvision
import mmdet
import mmcv
import os.path as osp

from mmcv.runner import load_checkpoint

from mmdet.apis import inference_detector, show_result_pyplot
from mmdet.models import build_detector
from mmdet.datasets import build_dataset
from mmdet.apis import train_detector
```

```python
# config = '/home/sjtu/scratch/tongzhao/mmdetection/configs/e2ec/e2ec_resnet18_dcnv2_coco.py'
config = '/home/sjtu/scratch/tongzhao/mmdetection/configs/centernet/centernet_resnet18_dcnv2_140e_coco.py'

device='cuda:0'

config = mmcv.Config.fromfile(config)

config.model.pretrained = None

# Build model
model = build_detector(config.model)
# Build dataset
datasets = [build_dataset(config.data.train)]
# Add an attribute for visualization convenience
model.CLASSES = datasets[0].CLASSES

model.cfg = config
model.to(device)
model.eval()
```

```python
# Create work_dir
# mmcv.mkdir_or_exist(osp.abspath(config.work_dir))
```

```python
train_detector(model, datasets, config, distributed=False, validate=True)
```
